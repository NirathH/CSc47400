# 2D Scalar Field

```python
display(Image(filename="Images/2D_ScalarField.png"))
```

This visualization is a clear example of a **2D scalar field**. The data represents a scalar quantity, like temperature or pressure, at every point in a 2D space. The color at each coordinate corresponds to the magnitude of that scalar value, with the highest values shown here along the central axis where r=0.

# 3D Scalar Field

```python
display(Image(filename="Images/3D_ScalarField.png"))
```

This visualization visualizes a scalar field using a **3D scalar field**. For every point on the 2D domain defined by the x and y axes, there is a corresponding scalar value which is represented in two ways: by its height along the z-axis and by its color. This could represent physical phenomena like the flow of a fluid, the velocity of wind in a space, or the strength and direction of a magnetic field.

# 2D Vector Field

```python
display(Image(filename="Images/2D_VectorField.png"))
```

This visualization is a **2D vector field** that consists of a vector (a value with both magnitude and direction) at every point on a two-dimensional plane. This could represent a physical force like a magnetic field or the velocity of a fluid across a surface. The visualization uses arrows to represent this data. The location of each arrow corresponds to a point on the x,y plane, its orientation shows the vector's direction, and its color indicates the vector's magnitude (strength or speed), as shown by the color bar on the right.

# 3D Shape

```python
display(Image(filename="Images/3D_Shape.png"))
```

This visualizaton is a **3D shape** that consists of a large set of vertices (the individual points) which are coordinates in 3D space. These vertices are connected by edges (the lines) to form a network of polygons that define the object's surface. This "wireframe" structure is a way to represent a 3D shape in computer graphics. Instead of being a solid volume or a field of data, it is a representation that describes the object's surface geometry, allowing it to be rendered, manipulated, and viewed from any angle.

# Graph

```python
display(Image(filename="Images/Graph.png"))
```

This visualization is a **graph** that is not a tree. The data consists of nodes (the subway stations) and edges (the rail lines connecting them). This map represents a graph because it models the relationships (connections) between a set of objects (stations). Importantly, it is not a tree because there are multiple paths between many nodes and numerous cycles (routes that loop back or form a closed circuit). This interconnectivity, which allows a traveler to take different routes to get to the same destination, is a key feature of a general graph.
